```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 300
m = []
for n in np.linspace(1,8,30):
    t = np.random.normal(1,1,size=int(10**n))
    m.append(t.mean())
m[0]
```

```python
plt.plot(m)
plt.ylabel('single sample mean')
plt.xlabel('10^n')
```
